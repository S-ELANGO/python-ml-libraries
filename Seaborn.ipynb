```python
#import seaborn library
import seaborn as sns
import matplotlib.pyplot as plt
#load the dataset default dataset is tips
df =sns.load_dataset('tips')
#check the dataset if loaded or not used head()
df.head()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df)
plt.title('Scatter plot')
plt.grid()
plt.show()
```

```python
#line plot KDE
sns.lineplot(x='size',y='total_bill',data=df)
plt.title("Line Plot")
plt.show()
```

```python
#Bar Plot
sns.barplot(x='day',y='total_bill',data=df)
plt.title("Bar Plot")
plt.show()
```

```python
#boxplot to find outliers
sns.boxplot(x='day',y='total_bill',data=df)
plt.title("Histogram")
plt.show()
```

```python
#histogram
sns.histplot(df['total_bill'],bins=10,kde=True)
plt.title("Histogram")
plt.show()
```

```python
#pairplot
sns.pairplot(df)
plt.show()
```

```python
#KDE plot find distribution
sns.kdeplot(df['total_bill'],shade=True)
plt.show()
```

```python
#correlation
corr=df[['total_bill','tip','size']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
